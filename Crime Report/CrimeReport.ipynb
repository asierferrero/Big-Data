```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import folium as fo
```

```python
lat = 34.05
lon = -118.2437
```

```python
data = pd.read_csv("Crime_Data_from_2010_to_2019.csv", delimiter=",")
```

```python
crimes = pd.DataFrame(data, columns=['LAT', 'LON', 'DATE OCC', 'VICT Sex', 'Crm Cd Desc', 'AREA NAME'])

crimes.rename(columns={'Crm Cd Desc' : 'Category'}, inplace=True)

limit = 100
crimes = crimes.iloc[0: limit]

crimes
```

```python
world_map = fo.Map(location=[lat, lon], zoom_start=16)

world_map
```

```python
from folium import plugins

incidents = fo.FeatureGroup()

for lat, lon in zip(crimes['LAT'], crimes['LON']):
    incidents.add_child(
        fo.CircleMarker(
            [lat, lon],
            radius=5,
            color="yellow",
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

latitudes = list(crimes['LAT'])
longitudes = list(crimes['LON'])
labels = list(crimes['Category'])

for lat, lon, label in zip(latitudes, longitudes, labels):
    fo.Marker([lat,lon], popup=label).add_to(world_map)

world_map.add_child(incidents)
    
```
